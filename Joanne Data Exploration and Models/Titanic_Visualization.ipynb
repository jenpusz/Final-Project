```python
#Import Dependencies

import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#kaggle_train = pd.read_csv('../DataSets/Kaggle/train.csv')
#kaggle_test = pd.read_csv('../DataSets/Kaggle/test.csv')
titanic_complete = pd.read_csv('../DataSets/TensorFlow/Titanic.csv')
```

```python
titanic_complete.shape
```

```python
# Dropping unwanted columns
titanic_complete=titanic_complete.drop(['name','cabin','body','boat','home.dest'],1)
titanic_complete
```

```python
sns.catplot(x = 'pclass', y='survived', kind='bar', data=titanic_complete)
```

```python
sns.catplot(x='pclass', y='survived', kind='violin', data=titanic_complete)
```

```python
sns.catplot(x='sex', y='survived', kind='bar', data=titanic_complete, hue='pclass')
```

```python
sns.catplot(x='sex', y='survived', kind='point', data=titanic_complete,hue='pclass')
```

```python
sns.countplot(x='sex', data=titanic_complete)
```

```python
sns.catplot(x='sex', y='survived',kind='bar',data=titanic_complete)
```

```python
sns.catplot(x='sex', y='survived', kind='violin', data=titanic_complete)
```
